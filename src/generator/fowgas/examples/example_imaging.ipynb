# Import stuff

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

from mysticetus import imaging
```

# `kwargs` for imshow

```python
iss = {'origin':'lower', 'interpolation':'nearest', 'cmap':plt.cm.CMRmap}
```

# Function examples

## center_and_rotate(array, center, angle)

```python
# image size
nx, ny = 32, 32

# index of pixel to rotate around
cx, cy = nx//2-5, ny//2

# rotation angle
angle = np.pi/4

# indices of white pixels
p = []
p.append((ny//2+3, nx//2+6-5))
p.append((ny//2-4, nx//2-4-5))
p.append((ny//2+3, nx//2+3-5))
p.append((ny//2-4, nx//2+3-5))
p.append((ny//2+3, nx//2-4-5))

# make original image
a = np.zeros(shape=(ny, nx))
for pp in p:
    a[pp] = 1

# center and rotate
b = imaging.center_and_rotate(a, (cy, cx), angle)



fig = plt.figure(figsize=(20,12))

plt.subplot(1, 2, 1)
d1  = a
im1 = plt.imshow(d1, **iss)

plt.subplot(1, 2, 2)
d2  = b
im2 = plt.imshow(d2, **iss)
```

## central_similar_portion_xyslice(array, factor)

```python
# make slice tuples
s1 = imaging.central_similar_portion_xyslice(a, 0.8)
s2 = imaging.central_similar_portion_xyslice(a, 0.25)



plt.figure(figsize=(20,12))

plt.subplot(1,2,1)
# use slice tuple
d1 = a[s1] 
plt.imshow(d1, **iss)

plt.subplot(1,2,2)
# use slice tuple
d2 = a[s2]
plt.imshow(d2, **iss)
```

## central_xyslice(array, ssize)

```python
# make slice tuples
s1 = imaging.central_xyslice(a, (16, 16))
s2 = imaging.central_xyslice(a, (16, 32))



plt.figure(figsize=(20,12))

plt.subplot(1,2,1)
# use slice tuple
d1 = a[s1]
plt.imshow(d1, **iss)

plt.subplot(1,2,2)
# use slice tuple
d2 = a[s2]
plt.imshow(d2, **iss)
```

## circular_increase_mask(mask, diameter)

```python
# make an increased mask
diameter = 5
c = imaging.circular_increase_mask(a, diameter)



plt.figure(figsize=(20,12))

plt.subplot(1,2,1)
d1 = a
plt.imshow(d1, **iss)

plt.subplot(1,2,2)
d2 = c
plt.imshow(d2, **iss)
```

## calc_angle(start, end)

```python
# calculate angle of a line segment from center to one point
angle = imaging.calc_angle((cy, cx), p[0])
# rotate by the negative angle plus 90 degrees
d = imaging.center_and_rotate(a, (cy, cx), -angle+np.pi/2)



plt.figure(figsize=(20,12))

plt.subplot(1,2,1)
d1 = a
plt.imshow(d1, **iss)

plt.subplot(1,2,2)
d2 = d
plt.imshow(d2, **iss)
```
