```python
# import packages
import numpy as np
from matplotlib import pyplot as plt
from scipy import special

# append path for loading own packages
import sys
sys.path.append("../packages")

# switch on inline plotting
%matplotlib inline
```

```python
# import own edges package
import edges

# plot some edge functions from the package
x = np.linspace(-1., 1., 1000, endpoint=True)

fig = plt.figure(figsize=(20, 6))

fig.add_subplot(1, 3, 1)
plt.plot(x, edges.erf95_edge(x), x, edges.erf99_edge(x), x, edges.erf999_edge(x))

fig.add_subplot(1, 3, 2)
plt.plot(x, edges.cos_edge(x))

fig.add_subplot(1, 3, 3)
plt.plot(x, edges.circular_edge(x, 0.2))
```

```python
fig = plt.figure(figsize=(20, 6))

elliptical_edge_1 = edges.make_elliptical_edge(0.5, 1.)
elliptical_edge_2 = edges.make_elliptical_edge(0.1, 1.)
elliptical_edge_3 = edges.make_elliptical_edge(0.4, 0.2)

sub1 = fig.add_subplot(1, 3, 1, adjustable='datalim', aspect=1.)
plt.plot(x, elliptical_edge_1(x))
#sub1.axes().set_aspect('equal', 'datalim')

fig.add_subplot(1, 3, 2, adjustable='datalim', aspect=1.)
plt.plot(x, elliptical_edge_2(x))
#plt.axes().set_aspect('equal', 'datalim')

fig.add_subplot(1, 3, 3, adjustable='datalim', aspect=1.)
plt.plot(x, 5.*elliptical_edge_3(x))
#plt.axes().set_aspect('equal', 'datalim')
```
