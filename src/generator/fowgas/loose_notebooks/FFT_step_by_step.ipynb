# DFT: Step by step

The calculation of a DFT is presented, showing the basic arithmetical operations that are behind the thing. The result is then compared to the one that is obtained by using an FFT library.

**The value of *tres* must never be different from 8. Otherwise this notebook will become inconsistent!**

## Import stuff

```python
import numpy as np
#from numpy import fft
from scipy import fftpack as fft
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# Import functions for making normal distributions
import sys
sys.path.append('../packages/')
from normal_distribution import *
```

## Define and plot normal distribution density function

```python
# Define normal distribution density function
sigma = 1.
mu    = 0.
fct_g = make_normal_distribution_density(sigma, mu, normalization=True, prefactor=1.)

# Define domain for sampling
tres = 8                              # <- DO NOT CHANGE THIS
tmin  = -5.*sigma
tmax  = -tmin * (tres - 2) / tres
t     = np.linspace(tmin, tmax, tres)

# Plot
plt.plot(t, fct_g(t))
```

## Calculate DFT step by step...
...with the formula given here: https://docs.scipy.org/doc/numpy/reference/routines.fft.html#module-numpy.fft which is:

\begin{equation}
    A_k = \sum_{m=0}^{n-1} a_m \, e^{-2 \pi i \frac{mk}{n}}
\end{equation}

```python
# Calculate sampling values in the time domain
a = fct_g(t)

# Initialize the frequency domain array
A = np.zeros(shape=a.shape, dtype=complex)


# Do the DFT calculation

# This is where the value of k goes which is connected to the frequency
# |                                |
# v                                v
A[0] =\
    a[0] * np.exp(-2.*np.pi*1j * 0*0/8) +\
    a[1] * np.exp(-2.*np.pi*1j * 1*0/8) +\
    a[2] * np.exp(-2.*np.pi*1j * 2*0/8) +\
    a[3] * np.exp(-2.*np.pi*1j * 3*0/8) +\
    a[4] * np.exp(-2.*np.pi*1j * 4*0/8) +\
    a[5] * np.exp(-2.*np.pi*1j * 5*0/8) +\
    a[6] * np.exp(-2.*np.pi*1j * 6*0/8) +\
    a[7] * np.exp(-2.*np.pi*1j * 7*0/8)
#     ^                          ^
#     |                          |
# This is where the value of m goes which is connected to the time

# The next frequency: k takes the next value
A[1] =\
    a[0] * np.exp(-2.*np.pi*1j * 0*1/8) +\
    a[1] * np.exp(-2.*np.pi*1j * 1*1/8) +\
    a[2] * np.exp(-2.*np.pi*1j * 2*1/8) +\
    a[3] * np.exp(-2.*np.pi*1j * 3*1/8) +\
    a[4] * np.exp(-2.*np.pi*1j * 4*1/8) +\
    a[5] * np.exp(-2.*np.pi*1j * 5*1/8) +\
    a[6] * np.exp(-2.*np.pi*1j * 6*1/8) +\
    a[7] * np.exp(-2.*np.pi*1j * 7*1/8)

A[2] =\
    a[0] * np.exp(-2.*np.pi*1j * 0*2/8) +\
    a[1] * np.exp(-2.*np.pi*1j * 1*2/8) +\
    a[2] * np.exp(-2.*np.pi*1j * 2*2/8) +\
    a[3] * np.exp(-2.*np.pi*1j * 3*2/8) +\
    a[4] * np.exp(-2.*np.pi*1j * 4*2/8) +\
    a[5] * np.exp(-2.*np.pi*1j * 5*2/8) +\
    a[6] * np.exp(-2.*np.pi*1j * 6*2/8) +\
    a[7] * np.exp(-2.*np.pi*1j * 7*2/8)

A[3] =\
    a[0] * np.exp(-2.*np.pi*1j * 0*3/8) +\
    a[1] * np.exp(-2.*np.pi*1j * 1*3/8) +\
    a[2] * np.exp(-2.*np.pi*1j * 2*3/8) +\
    a[3] * np.exp(-2.*np.pi*1j * 3*3/8) +\
    a[4] * np.exp(-2.*np.pi*1j * 4*3/8) +\
    a[5] * np.exp(-2.*np.pi*1j * 5*3/8) +\
    a[6] * np.exp(-2.*np.pi*1j * 6*3/8) +\
    a[7] * np.exp(-2.*np.pi*1j * 7*3/8)

A[4] =\
    a[0] * np.exp(-2.*np.pi*1j * 0*4/8) +\
    a[1] * np.exp(-2.*np.pi*1j * 1*4/8) +\
    a[2] * np.exp(-2.*np.pi*1j * 2*4/8) +\
    a[3] * np.exp(-2.*np.pi*1j * 3*4/8) +\
    a[4] * np.exp(-2.*np.pi*1j * 4*4/8) +\
    a[5] * np.exp(-2.*np.pi*1j * 5*4/8) +\
    a[6] * np.exp(-2.*np.pi*1j * 6*4/8) +\
    a[7] * np.exp(-2.*np.pi*1j * 7*4/8)

A[5] =\
    a[0] * np.exp(-2.*np.pi*1j * 0*5/8) +\
    a[1] * np.exp(-2.*np.pi*1j * 1*5/8) +\
    a[2] * np.exp(-2.*np.pi*1j * 2*5/8) +\
    a[3] * np.exp(-2.*np.pi*1j * 3*5/8) +\
    a[4] * np.exp(-2.*np.pi*1j * 4*5/8) +\
    a[5] * np.exp(-2.*np.pi*1j * 5*5/8) +\
    a[6] * np.exp(-2.*np.pi*1j * 6*5/8) +\
    a[7] * np.exp(-2.*np.pi*1j * 7*5/8)

A[6] =\
    a[0] * np.exp(-2.*np.pi*1j * 0*6/8) +\
    a[1] * np.exp(-2.*np.pi*1j * 1*6/8) +\
    a[2] * np.exp(-2.*np.pi*1j * 2*6/8) +\
    a[3] * np.exp(-2.*np.pi*1j * 3*6/8) +\
    a[4] * np.exp(-2.*np.pi*1j * 4*6/8) +\
    a[5] * np.exp(-2.*np.pi*1j * 5*6/8) +\
    a[6] * np.exp(-2.*np.pi*1j * 6*6/8) +\
    a[7] * np.exp(-2.*np.pi*1j * 7*6/8)

A[7] =\
    a[0] * np.exp(-2.*np.pi*1j * 0*7/8) +\
    a[1] * np.exp(-2.*np.pi*1j * 1*7/8) +\
    a[2] * np.exp(-2.*np.pi*1j * 2*7/8) +\
    a[3] * np.exp(-2.*np.pi*1j * 3*7/8) +\
    a[4] * np.exp(-2.*np.pi*1j * 4*7/8) +\
    a[5] * np.exp(-2.*np.pi*1j * 5*7/8) +\
    a[6] * np.exp(-2.*np.pi*1j * 6*7/8) +\
    a[7] * np.exp(-2.*np.pi*1j * 7*7/8)

print(A)
```

## Calculate DFT using the library FFT function

In fact this does something more clever then the rather lengthy calculation that is shown above. However, the results should be consistent with the above formula.

```python
A2 = fft.fft(a)

print(A2)
```

## Compare the results
As one can see, the absolute differences are in the range of numerical accuracy and can probably be accounted to different orderings of additions and the like.

```python
diff = np.abs(A2 - A)

print(diff)
```

## Plot the frequency spectrum

```python
# Calculate the sampling frequencies
f               = fft.fftfreq(tres, t[1]-t[0])

# Shift frequencies and spectrum samples to continuos ordering
f_shifted       = fft.fftshift(f)
A2_shifted      = fft.fftshift(A2)

# 
A2_shifted_real = np.real( A2_shifted)
A2_shifted_imag = np.imag( A2_shifted)
A2_shifted_abs  = np.abs(  A2_shifted)
A2_shifted_phs  = np.angle(A2_shifted) / np.pi         * np.max(A2_shifted_abs)
A2_shifted_pi   = np.ones(shape=A2_shifted_real.shape) * np.max(A2_shifted_abs)

# Plot
plt.plot(f_shifted, A2_shifted_real,       label='real part')
plt.plot(f_shifted, A2_shifted_imag,       label='imaginary part')
plt.plot(f_shifted, A2_shifted_abs,        label='absolute')
plt.plot(f_shifted, A2_shifted_phs,        label='phase (same unit as +/-pi)')
plt.plot(f_shifted, A2_shifted_pi,   '--', label='pi',)
plt.plot(f_shifted, -A2_shifted_pi,  '--', label='-pi')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

## Define and plot shifted normal distribution and its FFT

```python
# Define 
sigma = 1.
mu    = 0.
fct_g = make_normal_distribution_density(sigma, mu)

tres  = 32
tmin  = -5.*sigma
tmax  = -tmin * (tres-2) / tres
t     = np.linspace(tmin, tmax, tres)

temp  = fct_g(t)
g     = np.zeros(shape=temp.shape)
g[0:tres//2]    = temp[tres//2:tres]
g[tres//2:tres] = temp[0:tres//2]

G     = fft.fft(g)
G     = fft.fftshift(G)
f     = fft.fftfreq(tres, t[1]-t[0])
f     = fft.fftshift(f)

G_real = np.real( G)
G_imag = np.imag( G)
G_abs  = np.abs(  G)
G_phs  = np.angle(G) / np.pi        * np.max(G_abs)
G_pi   = np.ones(shape=G_abs.shape) * np.max(G_abs)

fig = plt.figure(figsize=(20,12))

sub1 = fig.add_subplot(2, 1, 1)
sub1.plot(t, g)
sub1.plot(t, temp, '--')

sub2 = fig.add_subplot(2, 1, 2)
sub2.plot(f, G_real,      label='real part')
sub2.plot(f, G_imag,      label='imaginary part')
sub2.plot(f, G_abs,       label='absolute')
sub2.plot(f, G_phs,       label='phase (same unit as +/-pi)')
sub2.plot(f, G_pi,  '--', label='pi')
sub2.plot(f, -G_pi, '--', label='-pi')
sub2.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```
