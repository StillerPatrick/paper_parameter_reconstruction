```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
from scipy import fftpack as fft
```

```python
import sys
sys.path.append('../packages')
from grating2 import *
from auxiliary import *
```

```python
# Define grating function
mu       = 0.
sigma    = 0.2
fct_edge = make_erf_edge(sigma, mu)

pitch    = 4.
fsize    = 0.8
fct_feat = make_box(pitch, fsize, fct_edge)

fct_grat = make_periodical(pitch, fct_feat, 0.)

# Define definition domain
xres     = 1024
xmin     = 0.
xmax     = 100.
x        = np.linspace(xmin, xmax, xres, endpoint=False)

# Plot
plt.plot(x, fct_grat(x))
```

```python
grat = fct_grat(x)

gratFFT = fft.fftshift(fft.fft(grat))
f = fft.fftshift(fft.fftfreq(len(x), x[1]-x[0]))


plot_complex(f[127:896], (gratFFT*f/np.exp(-4.*np.pi*f**2*sigma**2))[127:896], angl=False, real=False, imag=False)
```
