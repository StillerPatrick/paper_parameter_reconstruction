```python
# import packages
import numpy as np
from matplotlib import pyplot as plt
from scipy import fftpack

# append path for loading own packages
import sys
sys.path.append("../packages")

# switch on inline plotting
%matplotlib inline
```

```python
xmin = 0
xmax = 2.
xres = 256
x = np.linspace(xmin, xmax, xres, endpoint=True)
ymin = -1.
ymax = 2.
yres = 256
y = np.linspace(ymin, ymax, yres, endpoint=True)

import func2areamap
import grating
import auxiliary as aux

gr1 = func2areamap.func2areamap(x, y, grating.perfect_grating(1., 0.5)(x))
ft1 = fftpack.fft2(gr1)
ft1 = fftpack.fftshift(ft1)

gr2 = func2areamap.func2areamap(x, y, grating.perfect_grating(0.1, 0.05)(x))
ft2 = fftpack.fft2(gr2)
ft2 = fftpack.fftshift(ft2)

gr3 = func2areamap.func2areamap(x, y, grating.perfect_grating(0.01, 0.005)(x))
ft3 = fftpack.fft2(gr3)
ft3 = fftpack.fftshift(ft3)


fig = plt.figure(figsize=(20,20))

fig.add_subplot(3,3,1)
aux.imshow(aux.array2image(gr1))
plt.colorbar()
fig.add_subplot(3,3,4)
aux.imshow(aux.array2image(np.abs(ft1)))
fig.add_subplot(3,3,7)
aux.imshow(aux.logimage(aux.array2image(np.abs(ft1)**2)))

fig.add_subplot(3,3,2)
aux.imshow(aux.array2image(gr2))
plt.colorbar()
fig.add_subplot(3,3,5)
aux.imshow(aux.array2image(np.abs(ft2)))
fig.add_subplot(3,3,8)
aux.imshow(aux.logimage(aux.array2image(np.abs(ft2)**2)))

fig.add_subplot(3,3,3)
aux.imshow(aux.array2image(gr3))
plt.colorbar()
fig.add_subplot(3,3,6)
aux.imshow(aux.array2image(np.abs(ft3)))
fig.add_subplot(3,3,9)
aux.imshow(aux.logimage(aux.array2image(np.abs(ft3)**2)))

plt.show()
```
