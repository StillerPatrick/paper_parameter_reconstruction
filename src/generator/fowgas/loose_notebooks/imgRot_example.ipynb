```python
# import packages
import numpy as np
from matplotlib import pyplot as plt

# append path for loading own packages
import sys
sys.path.append("../packages")

# switch on inline plotting
%matplotlib inline
```

```python
# edge length of final array
n = 128

# import own rotation package
import imgRot

# create pre-rotation array - larger size!
li = imgRot.minInputLength(n)
arr = np.zeros((li,li))

# set some array values
arr[0:li, 0:li/2] = np.ones((li,li))[0:li, 0:li/2]

# show pre-rotation array
plt.imshow(np.swapaxes(arr,0,1), interpolation='nearest', origin='lower', cmap=plt.cm.gray)
plt.colorbar()
plt.show()
```

```python
# show rotated arrays - correct size n!
fig = plt.figure(figsize=(20,12))

angles = [0., 3., 20., 45., 90., 117., 180., 270., 303.]
for i in range(9):
    angle = angles[i]
    a = imgRot.cropRotate(arr, n, angle)
    fig.add_subplot(3, 3, i+1)
    plt.imshow(np.swapaxes(a,0,1), interpolation='nearest', origin='lower', cmap=plt.cm.gray)
    plt.colorbar()

plt.show()
    
```
