```python
# import common packages
import numpy as np
from matplotlib import pyplot as plt

# append path for loading own packages
import sys
sys.path.append("../packages")

# switch on inline plotting
%matplotlib inline
```

```python
# definition of domain
xmin = 0.
xmax = 100.
xlength = xmax - xmin
```

```python
# import own grating package
import grating

# create three perfect grating functions
pitch = 0.10*xlength
feature = 0.5*pitch
f1 = grating.perfect_grating(pitch, feature)

feature = 0.25*pitch
f2 = grating.perfect_grating(pitch, feature)

pitch = 0.25*xlength
feature = 0.25*pitch
f3 = grating.perfect_grating(pitch, feature)
```

```python
# prepare and plot the perfect grating functions
xres = 256
x = np.linspace(xmin, xmax, xres, endpoint='True')

fig = plt.figure(figsize=(20,6))

fig.add_subplot(1, 3, 1)
plt.plot(x, f1(x))
plt.ylim(-0.5, 1.5)

fig.add_subplot(1, 3, 2)
plt.plot(x, f2(x))
plt.ylim(-0.5, 1.5)

fig.add_subplot(1, 3, 3)
plt.plot(x, f3(x))
plt.ylim(-0.5, 1.5)

plt.show()
```

```python
# define some feature function
gf = np.vectorize(lambda x: 0. if np.abs(x)>=10. else (1.+0.1*np.cos(2.*x)))

# make grating from that feature function
g  = grating.feature_grating(30., gf)

# show
plt.plot(x, g(x))
```

```python
# define some edge function
ee = np.vectorize(lambda x: 0.5*(1.+np.sign(x)) if np.abs(x)>0.5 else x+0.5)

# make feature from edge
ef = grating.symm_feat(2., ee)

# make grating from feature
eg = grating.feature_grating(5., ef)

# show
ex = np.linspace(-10., 10., 512, endpoint=True)
fig = plt.figure(figsize=(20,6))

fig.add_subplot(1,2,1)
plt.plot(ex, ef(ex))

fig.add_subplot(1,2,2)
plt.plot(ex, eg(ex))
```
