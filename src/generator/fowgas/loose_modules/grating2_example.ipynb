```python
# Example: make_periodical
sigma      = 1.
mu         = 3.
pitch      = 6.
offset     = 0.

xmin       = -20.
xmax       = -xmin
xres       = 256
x          = np.linspace(xmin, xmax, xres)

fct_normal = make_normal_distribution_density(sigma, mu)
fct_period = grating2.make_periodical(pitch, fct_normal, offset)

g          = fct_period(x)

plt.plot(x, g)
```

```python
# Example: make_box
sigma    = 1.
mu       = 0.
fsize    = 5.
pitch    = 13.

fct_edge = grating2.make_erf_edge(sigma, mu)
fct_box  = grating2.make_box(pitch, fsize, fct_edge)

xres     = 128
xmin     = -1.
xmax     = pitch + 1.
x        = np.linspace(xmin, xmax, xres)

g = fct_box(x)

plt.plot(x, g)
```
